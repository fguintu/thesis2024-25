```python
import h5py
import pandas as pd

# Open the HDF5 file in read mode
h5 = h5py.File("UP000000589_10090per-protein.h5", 'r')
# print filename
print(h5.filename)

# Assuming the file contains a dataset that you want to load into a DataFrame
# Replace 'dataset_name' with the actual name of the dataset you want to load
dataset_name = list(h5.keys())[0]  # Access the first dataset or specify the exact name
print(list(h5.keys()))
print(len(list(h5.keys())))
print(dataset_name)
data = h5[dataset_name][:]  # Load the dataset into a NumPy array

# If your dataset contains tabular data (e.g., 2D array), convert it to a DataFrame
df = pd.DataFrame(data)

# If the dataset has named columns (e.g., stored as attributes), you can set them as column names
if 'columns' in h5[dataset_name].attrs:
    df.columns = h5[dataset_name].attrs['columns']

# Print the first few rows of the DataFrame
print(df.shape)

# Close the file
h5.close()
```
