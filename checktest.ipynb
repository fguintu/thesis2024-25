```python
import h5py

# Open the HDF5 file
with h5py.File('test.h5', 'r') as f:
    # Print all root level object names (aka keys) 
    print("Keys: %s" % f.keys())
    
    # Get the first object name/key
    a_group_key = list(f.keys())[0]

    # Get the object type
    print(type(f[a_group_key]))  # <class 'h5py._hl.group.Group'>

    print(f[a_group_key])
    # Get the data
    data = f[a_group_key][:]
    print(data.shape)
```
